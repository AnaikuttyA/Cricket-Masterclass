```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.expand_frame_repr',False)
pd.set_option('max_colwidth',None)
```

```python
ipl_df = pd.read_csv(r"C:\Users\ANAIKUTTY\Documents\Cricket analysis\data files\ipl_2008_2024_ball_by_ball_data.csv")
```

```python
df = ipl_df.copy()
```

```python
df['phase'] = df['ball'].apply(lambda x:'powerplay' if x < 6 else ( 'middle' if x < 15 else 'death over'))
```

```python
df['isdot'] = df['runs_off_bat'].apply(lambda x: 1 if x == 0 else 0)
df['isone'] = df['runs_off_bat'].apply(lambda x: 1 if x == 1 else 0)
df['istwo'] = df['runs_off_bat'].apply(lambda x: 1 if x == 2 else 0)
df['isthree'] = df['runs_off_bat'].apply(lambda x: 1 if x == 3 else 0)
df['isfour'] = df['runs_off_bat'].apply(lambda x: 1 if x == 4 else 0)
df['issix'] = df['runs_off_bat'].apply(lambda x: 1 if x == 6 else 0)
```

```python
pp_df = df[df['phase']== 'powerplay']
```

```python
pp_df.head(1)
```

```python
pp_df = pd.DataFrame(pp_df.groupby('striker').agg({'match_id':'nunique','runs_off_bat':'sum','ball':'count','player_dismissed':'count','isfour':'sum','issix':'sum','isdot':'sum'})).reset_index()
pp_df = pp_df.rename(columns={'match_id':'matches','runs_off_bat':'runs','ball':'balls','player_dismissed':'outs','isfour':'four','issix':'six','isdot':'dots'})
```

```python
pp_df['strikerate'] = pp_df['runs'] / pp_df['balls'] * 100
pp_df['average'] = np.where(pp_df['outs']==0,pp_df['runs'],pp_df['runs'] / pp_df['outs'])
pp_df['boundary_%']  = ((pp_df['four']*4 + pp_df['six']* 6) / pp_df['runs']) *100
pp_df['dot_%'] = pp_df['dots'] / pp_df['balls'] *100
```

```python
pp_df = pp_df[pp_df['matches'] >= 20].sort_values(by='strikerate',ascending=False)
```
